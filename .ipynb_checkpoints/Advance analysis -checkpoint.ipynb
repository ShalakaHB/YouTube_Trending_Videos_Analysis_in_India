# Category performance analysis

## Time-Series Analysis for Views and Likes over Time

```python
import pandas as pd
trending_videos_IN = pd.read_csv('trending_videos_IN.csv')
# Convert 'published_at' to datetime
trending_videos_IN['published_at'] = pd.to_datetime(trending_videos_IN['published_at'])

# Group by date (ignoring time) and calculate total views and likes
time_series_data = trending_videos_IN.groupby(trending_videos_IN['published_at'].dt.date).agg({
    'view_count': 'sum',
    'like_count': 'sum',
    'comment_count': 'sum'
}).reset_index()

# Plotting time-series data for views and likes
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(time_series_data['published_at'], time_series_data['view_count'], label='Total Views', color='blue')
plt.plot(time_series_data['published_at'], time_series_data['like_count'], label='Total Likes', color='green')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Views and Likes Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Category Performance Analysis

```python
import pandas as pd

# Assuming the dataset is loaded in trending_videos_IN
# Create the engagement metrics columns
trending_videos_IN['likes_to_views'] = trending_videos_IN['like_count'] / trending_videos_IN['view_count']
trending_videos_IN['comments_to_views'] = trending_videos_IN['comment_count'] / trending_videos_IN['view_count']

# Replace infinities or NaN with 0 for videos with zero views
trending_videos_IN.fillna(0, inplace=True)

# Group by 'category_id' and calculate mean values for views, likes, comments, and engagement metrics
category_performance = trending_videos_IN.groupby('category_id').agg({
    'view_count': 'mean',
    'like_count': 'mean',
    'comment_count': 'mean',
    'likes_to_views': 'mean',
    'comments_to_views': 'mean'
}).reset_index()

# Sort by average view count to see the best performing categories
category_performance = category_performance.sort_values(by='view_count', ascending=False)

# Display the category performance metrics
print(category_performance)
```

## Observations:
- Category 28 has the highest average views (3.17M) with solid engagement (4.37% likes-to-views and 0.33% comments-to-views).
- Category 27 stands out for the highest engagement with a 13% likes-to-views ratio and 1.08% comments-to-views ratio, despite having lower average views (323K).
- Categories 23 and 20 also have high engagement rates regarding likes-to-views ratio.
- Category 2 has low engagement and relatively few views compared to other categories.




This analysis gives insights into how different video categories perform in terms of popularity (views) and engagement (likes/comments).

## Map Category Names

```python
# Add category names to the dataframe using the mapping dictionary
category_mapping = {
    1: "Film & Animation",
    2: "Autos & Vehicles",
    10: "Music",
    17: "Sports",
    20: "Gaming",
    22: "People & Blogs",
    23: "Comedy",
    24: "Entertainment",
    25: "News & Politics",
    26: "How-to & Style",
    27: "Education",
    28: "Science & Technology"
}

# Map the category_id to category names
category_performance['category_name'] = category_performance['category_id'].map(category_mapping)

# Reorder the columns for better readability
category_performance = category_performance[['category_name', 'view_count', 'like_count', 'comment_count', 'likes_to_views', 'comments_to_views']]

# Display the category performance with names
print(category_performance)
```

## Insights:
- Science & Technology has the highest average views, making it the most popular category.
- Education has the highest engagement rates, with 13.04% likes-to-views and 1.08% comments-to-views, even though it has lower average views.
- Comedy and Gaming categories also show strong engagement ratios in likes-to-views.
- Autos & Vehicles has the lowest engagement and views.


These insights help in understanding the popularity and engagement across various YouTube categories in India.

## Visualize category trends

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size for all plots
plt.figure(figsize=(14, 10))

# Bar plot for average views, likes, and comments across categories
plt.subplot(3, 1, 1)
sns.barplot(x='view_count', y='category_name', data=category_performance, palette='viridis')
plt.title('Average Views by Category')
plt.xlabel('Average View Count')
plt.ylabel('Category')

plt.subplot(3, 1, 2)
sns.barplot(x='like_count', y='category_name', data=category_performance, palette='magma')
plt.title('Average Likes by Category')
plt.xlabel('Average Like Count')
plt.ylabel('Category')

plt.subplot(3, 1, 3)
sns.barplot(x='comment_count', y='category_name', data=category_performance, palette='coolwarm')
plt.title('Average Comments by Category')
plt.xlabel('Average Comment Count')
plt.ylabel('Category')

plt.tight_layout()
plt.show()

# Scatter plot for engagement metrics (likes-to-views and comments-to-views)
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x='likes_to_views', y='category_name', data=category_performance, hue='likes_to_views', size='view_count', palette='Blues', sizes=(50, 500))
plt.title('Likes-to-Views Ratio by Category')
plt.xlabel('Likes to Views Ratio')
plt.ylabel('Category')

plt.subplot(1, 2, 2)
sns.scatterplot(x='comments_to_views', y='category_name', data=category_performance, hue='comments_to_views', size='view_count', palette='Greens', sizes=(50, 500))
plt.title('Comments-to-Views Ratio by Category')
plt.xlabel('Comments to Views Ratio')
plt.ylabel('Category')

plt.tight_layout()
plt.show()

# Stacked bar chart for views, likes, and comments to show the contribution of engagement metrics per category
plt.figure(figsize=(12, 6))

# Normalize engagement metrics to sum to 1 for each category
category_performance['total_engagement'] = category_performance['like_count'] + category_performance['comment_count']
category_performance['like_ratio'] = category_performance['like_count'] / category_performance['total_engagement']
category_performance['comment_ratio'] = category_performance['comment_count'] / category_performance['total_engagement']

# Stacked bar chart
category_performance[['category_name', 'like_ratio', 'comment_ratio']].set_index('category_name').plot(
    kind='bar', stacked=True, figsize=(12, 6), colormap='Spectral')

plt.title('Proportion of Likes and Comments Engagement by Category')
plt.xlabel('Category')
plt.ylabel('Proportion of Engagement')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

## Explanation of Visualizations:
- Bar Plot: Visualizes the average number of views, likes, and comments for each category.
- Scatter Plot: Shows the relationship between engagement metrics (likes-to-views and comments-to-views) and the category. The size of each point represents the average views in that category.
- Stacked Bar Chart: Displays the proportion of likes and comments as part of the total engagement, allowing you to see how much of the engagement comes from likes vs. comments in each category.


These visualizations help provide deeper insights into how each category performs in terms of views and engagement.
