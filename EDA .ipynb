### 5. Analyzing Numeric Features

```python
# Visualizing the distribution of views, likes, dislikes, and comment count
fig, ax = plt.subplots(2, 2, figsize=(14, 10))

# Distribution of Views
sns.histplot(df_cleaned['view_count'], bins=30, ax=ax[0, 0])
ax[0, 0].set_title('Distribution of Views')

# Distribution of Likes
sns.histplot(df_cleaned['like_count'], bins=30, ax=ax[0, 1])
ax[0, 1].set_title('Distribution of Likes')

# Distribution of Dislikes
sns.histplot(df_cleaned['dislike_count'], bins=30, ax=ax[1, 0])
ax[1, 0].set_title('Distribution of Dislikes')

# Distribution of Comment Count
sns.histplot(df_cleaned['comment_count'], bins=30, ax=ax[1, 1])
ax[1, 1].set_title('Distribution of Comment Count')

plt.tight_layout()
plt.show()
```

### 6. Correlation Analysis

```python
# correlation matrix
correlation_matrix = df[['view_count', 'like_count', 'comment_count']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='black')
plt.title('Correlation Matrix of Engagement Metrics')
plt.show()
```

```python
from textblob import TextBlob

# Function to calculate sentiment polarity
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

# Apply the function to the video titles or descriptions
df_cleaned['sentiment'] = df_cleaned['title'].apply(get_sentiment)

# Distribution of sentiment scores
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['sentiment'], bins=20)
plt.title('Distribution of Sentiment Scores')
plt.show()
```

### 9. Time-Based Analysis

```python
# Convert 'published_at' to datetime format
df_cleaned['published_at'] = pd.to_datetime(df_cleaned['published_at'])

# Plot number of trending videos over time
df_cleaned['publish_date'] = df_cleaned['published_at'].dt.date

# Visualizing the number of trending videos over time
plt.figure(figsize=(10, 6))
df_cleaned['publish_date'].value_counts().sort_index().plot()
plt.title('Number of Trending Videos Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Trending Videos')
plt.show()
```
